```python

#!/usr/bin/env python3
from pathlib import Path
import shutil

DATASET_NAME = f"Dual_Generator_Face_Reen" #f"FastReen" #f"FSRT"  #f"DCFace" #f"Face2Diffusion"

# Assume these variables are already set (from args)
dataset_root = f"/content/drive/MyDrive/AI_or_Real_datasets/FairFD/FairFD/dataset/test/{DATASET_NAME}"
dest_root = f"/content/drive/MyDrive/AI_or_Real_datasets/FairFD_Indian/fake"
race = "Indian"
copy = True
overwrite = False
remove_empty = False
dry_run = False

IMAGE_EXTS = {'.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.tif', '.webp'}

dataset_root = Path(dataset_root)
dest_root = Path(dest_root)
race_dir = dataset_root / race
dest_root.mkdir(parents=True, exist_ok=True)
dataset_name = dataset_root.name

for vid in sorted(p for p in race_dir.iterdir() if p.is_dir()):
    new_folder = dest_root / f"{dataset_name}_{race}_{vid.name}"
    new_folder.mkdir(parents=True, exist_ok=True)
    moved = 0
    for f in sorted(vid.iterdir()):
        if not f.is_file() or f.suffix.lower() not in IMAGE_EXTS:
            continue
        target = new_folder / f.name
        if target.exists() and not overwrite:
            i = 1
            while True:
                candidate = new_folder / f"{f.stem}_{i}{f.suffix}"
                if not candidate.exists():
                    target = candidate
                    break
                i += 1
        if dry_run:
            print(f"{'COPY' if copy else 'MOVE'}: {f} -> {target}")
        else:
            if copy:
                shutil.copy2(f, target)
            else:
                shutil.move(str(f), str(target))
        moved += 1
    if remove_empty and not dry_run:
        try:
            # remove vid dir if empty
            next(vid.iterdir())
        except StopIteration:
            vid.rmdir()
    print(f"Processed {vid.name}: moved/copied {moved} files to {new_folder}")
```

```python
#!/usr/bin/env python3
"""
Simple utility to explore the contents of a ZIP file and print a tree-like view.

Usage:
    python explore_zip.py /path/to/archive.zip

Output shows directories and files with compressed/uncompressed sizes.
"""
import argparse
import zipfile
from collections import defaultdict
from pathlib import PurePosixPath

def build_tree(infolist):
    tree = {}
    for info in infolist:
        path = PurePosixPath(info.filename)
        parts = path.parts
        node = tree
        for p in parts[:-1]:
            node = node.setdefault(p + "/", {})
        # last part: file or directory
        last = parts[-1]
        if info.is_dir() or info.filename.endswith("/"):
            node.setdefault(last + "/", {})
        else:
            node[last] = (info.compress_size, info.file_size)
    return tree

def print_tree(node, indent=""):
    items = sorted(node.items(), key=lambda x: (not x[0].endswith("/"), x[0].lower()))
    for name, val in items:
        if isinstance(val, dict):
            print(f"{indent}{name}")
            print_tree(val, indent + "    ")
        else:
            comp, orig = val
            print(f"{indent}{name}    ({orig} bytes, compressed {comp} bytes)")


zip_filepath = f"/content/drive/MyDrive/AI_or_Real_datasets/HIDF_dataset/Real-img.zip"
with zipfile.ZipFile(zip_filepath, "r") as z:
    #print(z.namelist())
    infos = z.infolist()
    tree = build_tree(infos)
    print(f"Archive: {zip_filepath}")
    print_tree(tree)
```

```python
import os
import glob
import shutil

root_folder = "/content/drive/MyDrive/AI_or_Real_datasets/HIDF_dataset_Indian/videos_indian_filtered/fake/frames" # Replace with your root folder path

for video_folder in os.listdir(root_folder):
    video_folder_path = os.path.join(root_folder, video_folder)

    # Check if it's a directory
    if os.path.isdir(video_folder_path):
        # Find all image files in the folder
        image_files = glob.glob(os.path.join(video_folder_path, "*.jpg")) + \
                      glob.glob(os.path.join(video_folder_path, "*.png")) + \
                      glob.glob(os.path.join(video_folder_path, "*.jpeg"))

        if image_files:
            # Sort the image files lexicographically and get the smallest
            smallest_image_path = sorted(image_files)[0]
            smallest_image_name = os.path.basename(smallest_image_path)
            smallest_image_ext = os.path.splitext(smallest_image_name)[1]

            # Construct the new name for the smallest image
            new_image_name = f"{video_folder}_{os.path.splitext(smallest_image_name)[0]}{smallest_image_ext}"
            new_image_path = os.path.join(root_folder, new_image_name)

            # Move and rename the smallest image to the root folder
            shutil.move(smallest_image_path, new_image_path)
            print(f"Moved and renamed: {smallest_image_path} to {new_image_path}")

            # Remove the original video folder and its contents
            shutil.rmtree(video_folder_path)
            print(f"Removed original folder: {video_folder_path}")
        else:
            print(f"No images found in {video_folder_path}. Removing empty folder.")
            shutil.rmtree(video_folder_path)
            print(f"Removed empty folder: {video_folder_path}")
```
